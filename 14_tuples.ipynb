# What are Tuples?

A tuple is a **collection of items stored in a single variable**.

Tuples are:

* Ordered
* Immutable (cannot be changed)
* Allow duplicate values

```python
numbers = (1, 2, 3, 4)

print(numbers)
```

# Creating Tuples

```python
t1 = (1, 2, 3)

# Single element tuple
t2 = (5,)

print(t1)
print(t2)
```

# Indexing in Tuples

```python
t = (10, 20, 30, 40)

print(t[0])
print(t[2])
```

# Negative Indexing

```python
t = (10, 20, 30, 40)

print(t[-1])
print(t[-2])
```

# Tuple Slicing

```python
t = (1, 2, 3, 4, 5)

print(t[1:4])
print(t[:3])
print(t[2:])
```

# Tuple Properties

* Tuples are immutable (cannot be changed)
* Faster than lists
* Useful for fixed data

```python
t = (1, 2, 3)

# This will give error
# t[0] = 10
```

# Tuple Methods

Only two main methods:

* count()
* index()

```python
t = (1, 2, 2, 3)

print(t.count(2))
print(t.index(3))
```

# Practice Problems

```python
# Problem 1: Print all elements of tuple

t = (1, 2, 3, 4)

for item in t:
    print(item)
```

```python
# Problem 2: Count occurrences of an element

t = (1, 2, 2, 3, 2)

print(t.count(2))
```

```python
# Problem 3: Find index of element

t = (10, 20, 30)

print(t.index(20))
```

```python
# Problem 4: Convert tuple to list

t = (1, 2, 3)

lst = list(t)
print(lst)
```
