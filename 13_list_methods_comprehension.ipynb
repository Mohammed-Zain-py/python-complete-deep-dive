# What are List Methods?

List methods are built-in functions used to **modify and work with lists**.

# Adding Elements

```python
numbers = [1, 2, 3]

numbers.append(4)     # add at end
print(numbers)

numbers.insert(1, 10)  # insert at index
print(numbers)
```

# Removing Elements

```python
numbers = [1, 2, 3, 4, 5]

numbers.remove(3)   # remove value
print(numbers)

numbers.pop()       # remove last element
print(numbers)
```

# Other Useful Methods

```python
numbers = [1, 2, 2, 3, 4]

print(numbers.count(2))   # count occurrences
print(numbers.index(3))   # find index
```

```python
numbers = [5, 2, 8, 1]

numbers.sort()
print(numbers)

numbers.reverse()
print(numbers)
```

# Copying Lists

```python
a = [1, 2, 3]

b = a.copy()

print(b)
```

# List Comprehension

List comprehension provides a **short and clean way to create lists**.

Syntax:

[expression for item in iterable]

```python
# Create list of squares

squares = [i*i for i in range(5)]

print(squares)
```

```python
# With condition

even_numbers = [i for i in range(10) if i % 2 == 0]

print(even_numbers)
```

# Practice Problems

```python
# Problem 1: Add element to list

numbers = [1, 2, 3]
numbers.append(4)

print(numbers)
```

```python
# Problem 2: Remove element from list

numbers = [1, 2, 3, 4]
numbers.remove(2)

print(numbers)
```

```python
# Problem 3: Sort a list

numbers = [5, 1, 3, 2]
numbers.sort()

print(numbers)
```

```python
# Problem 4: Create list of even numbers using comprehension

even = [i for i in range(1, 21) if i % 2 == 0]

print(even)
```

```python
# Problem 5: Create list of squares

squares = [i*i for i in range(1, 6)]

print(squares)
```
