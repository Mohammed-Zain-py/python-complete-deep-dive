# What are Function Arguments?

Arguments are values passed to a function when calling it.

They allow functions to work with different inputs.

# Positional Arguments

Values are passed in the same order as parameters.

```python
def add(a, b):
    print(a + b)

add(5, 3)
```

# Keyword Arguments

We can pass arguments using parameter names.

```python
def greet(name, age):
    print(name, age)

greet(age=21, name="Zain")
```

# Default Arguments

We can assign default values to parameters.

```python
def greet(name="User"):
    print("Hello", name)

greet()
greet("Zain")
```

# Variable-Length Arguments (*args)

Used when we don’t know how many arguments will be passed.

```python
def add(*numbers):
    total = 0
    for num in numbers:
        total += num
    print(total)

add(1, 2, 3, 4)
```

# Keyword Variable-Length Arguments (**kwargs)

Used to pass multiple keyword arguments.

```python
def display(**data):
    for key, value in data.items():
        print(key, ":", value)

display(name="Zain", age=21)
```

# Practice Problems

```python
# Problem 1: Function with default argument

def power(num, exponent=2):
    return num ** exponent

print(power(5))
print(power(5, 3))
```

```python
# Problem 2: Sum using *args

def total_sum(*nums):
    return sum(nums)

print(total_sum(1, 2, 3, 4, 5))
```

```python
# Problem 3: Display user info using **kwargs

def show_info(**info):
    for key, value in info.items():
        print(key, ":", value)

show_info(name="Zain", age=21, city="Bangalore")
```

```python
# Problem 4: Function using keyword arguments

def student(name, marks):
    print(name, marks)

student(marks=95, name="Zain")
```
