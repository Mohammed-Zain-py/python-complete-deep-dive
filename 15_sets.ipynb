# What are Sets?

A set is a **collection of unique elements**.

Sets are:

* Unordered
* Do not allow duplicates
* Mutable

```python
numbers = {1, 2, 3, 4}

print(numbers)
```

# Creating Sets

```python
s = {1, 2, 3}

# Creating empty set
empty_set = set()

print(s)
print(empty_set)
```

# Properties of Sets

```python
s = {1, 2, 2, 3, 3, 4}

print(s)   # duplicates removed
```

# Adding and Removing Elements

```python
s = {1, 2, 3}

s.add(4)
print(s)

s.remove(2)
print(s)
```

# Set Operations

```python
a = {1, 2, 3}
b = {3, 4, 5}

# Union
print(a | b)

# Intersection
print(a & b)

# Difference
print(a - b)
```

# Set Methods

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
```

# Practice Problems

```python
# Problem 1: Remove duplicates from list

lst = [1, 2, 2, 3, 3, 4]

unique = set(lst)

print(unique)
```

```python
# Problem 2: Find common elements

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]

print(set(a) & set(b))
```

```python
# Problem 3: Check if element exists

s = {1, 2, 3}

print(2 in s)
print(5 in s)
```

```python
# Problem 4: Union of two sets

a = {1, 2}
b = {3, 4}

print(a.union(b))
```

```python
# Problem 5: Difference of sets

a = {1, 2, 3}
b = {2, 3}

print(a - b)
```
