# What are Variables?

A variable is a **container used to store data** in a program.

In Python, we do not need to declare the type of a variable explicitly.

Example:

```python
x = 10
name = "Zain"
price = 99.99

print(x)
print(name)
print(price)
```

# Rules for Naming Variables

While naming variables, we should follow these rules:

* Variable names can contain letters, digits, and underscores
* They must start with a letter or underscore
* They cannot start with a number
* They are case-sensitive

Valid examples:

```python
my_variable = 10
_name = "Python"
age2 = 21
```

```python
# Invalid examples (will cause error)

# 2name = "Zain"
# my-variable = 10
```

# Data Types in Python

Python has different types of data:

* int → Integer numbers
* float → Decimal numbers
* str → String (text)
* bool → True or False
* list → Collection of items
* tuple → Immutable collection
* dict → Key-value pairs

```python
a = 10          # int
b = 10.5        # float
c = "Hello"     # string
d = True        # boolean

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

# Type Checking

We use the `type()` function to check the data type of a variable.

```python
x = 100
print(type(x))
```

# Type Casting

Type casting means **converting one data type into another**.

There are two types:

* Implicit Type Casting (automatic)
* Explicit Type Casting (manual)

# Implicit Type Casting

Python automatically converts data types when needed.

```python
a = 5       # int
b = 2.5     # float

c = a + b   # int + float → float

print(c)
print(type(c))
```

# Explicit Type Casting

We manually convert data types using functions like:

* int()
* float()
* str()

```python
x = "10"

# Convert string to integer
y = int(x)

print(y)
print(type(y))
```

```python
a = 5
b = float(a)

print(b)
print(type(b))
```
