# What are Functions?

A function is a **block of code that performs a specific task**.

Functions help in:

* Reusing code
* Making programs organized
* Improving readability

# Defining a Function

We use the `def` keyword to define a function.

Syntax:

def function_name():
    code

```python
def greet():
    print("Hello, welcome to Python")

greet()
```

# Function with Parameters

Parameters allow us to pass values to a function.

```python
def greet(name):
    print("Hello", name)

greet("Zain")
```

# Multiple Parameters

```python
def add(a, b):
    print(a + b)

add(5, 3)
```

# Return Statement

The `return` statement is used to send a value back from the function.

```python
def add(a, b):
    return a + b

result = add(10, 5)
print(result)
```


# Difference Between print() and return

print() → Displays output  
return → Sends value back to caller

```python
def test():
    return 10

x = test()
print(x)
```

# Practice Problems

```python
# Problem 1: Function to print square of a number

def square(num):
    print(num * num)

square(5)
```

```python
# Problem 2: Function to check even or odd

def check_even(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")

check_even(7)
```

```python
# Problem 3: Function to return sum of two numbers

def add(a, b):
    return a + b

print(add(4, 6))
```

```python
# Problem 4: Function to find maximum of two numbers

def maximum(a, b):
    if a > b:
        return a
    else:
        return b

print(maximum(10, 20))
```
