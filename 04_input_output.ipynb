# Taking User Input

In Python, we use the `input()` function to take input from the user.

By default, input is stored as a **string**.

```python
name = input("Enter your name: ")
print("Hello", name)
```

# Important Note

The `input()` function always returns a string.

So even if you enter a number, Python treats it as a string.

```python
num = input("Enter a number: ")

print(num)
print(type(num))  # Always string
```

# Type Conversion with Input

To perform calculations, we must convert input into the required type.

Common conversions:

* int()
* float()
* str()

```python
num = input("Enter a number: ")

num = int(num)

print(num + 5)
```

```python
price = input("Enter price: ")

price = float(price)

print(price + 10.5)
```

# Multiple Inputs

We can take multiple inputs using split().

```python
a, b = input("Enter two numbers: ").split()

a = int(a)
b = int(b)

print("Sum:", a + b)
```

# Practice Problems

```python
# Problem 1: Take a number and print its square

num = int(input("Enter a number: "))
print("Square:", num * num)
```

```python
# Problem 2: Take two numbers and print their sum

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
```

```python
# Problem 3: Take age as input and print after 5 years

age = int(input("Enter your age: "))
print("Age after 5 years:", age + 5)
```

```python
# Problem 4: Take two numbers and print their product

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Product:", a * b)
```
