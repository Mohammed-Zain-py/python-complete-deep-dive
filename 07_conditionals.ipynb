# What are Conditional Statements?

Conditional statements are used to **make decisions in a program**.

They allow the program to execute different blocks of code based on conditions.

Example:

If a number is greater than 10 → do something  
Else → do something else

# if Statement

The `if` statement is used to check a condition.

Syntax:

if condition:
    code

```python
num = 15

if num > 10:
    print("Number is greater than 10")
```

# if-else Statement

Used when we have two possible outcomes.

```python
num = 5

if num > 10:
    print("Greater than 10")
else:
    print("Less than or equal to 10")
```

# elif Statement

Used when we have multiple conditions.

```python
num = 0

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

# Nested Conditions

We can use if statements inside another if statement.

```python
num = 15

if num > 0:
    if num % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
```

# Practice Problems

```python
# Problem 1: Check if a number is even or odd

num = 8

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Problem 2: Check if a number is positive, negative, or zero

num = -3

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Problem 3: Find the largest of two numbers

a = 10
b = 20

if a > b:
    print("a is greater")
else:
    print("b is greater")
```

```python
# Problem 4: Check eligibility for voting (age >= 18)

age = 17

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")
```
