# What are Strings?

A string is a **sequence of characters** enclosed in quotes.

We can use:

* Single quotes (' ')
* Double quotes (" ")
* Triple quotes (''' ''' or """ """)

```python
name = "Zain"
message = 'Hello World'
multi_line = """This is
a multi-line string"""

print(name)
print(message)
print(multi_line)
```

# String Indexing

Each character in a string has an index.

Index starts from 0.

Negative indexing starts from -1.

```python
text = "Python"

print(text[0])   # P
print(text[1])   # y
print(text[-1])  # n
print(text[-2])  # o
```

# String Slicing

Slicing is used to extract a part of the string.

Syntax:

string [start : end : step]

```python
text = "Python"

print(text[0:4])   # Pyth
print(text[1:5])   # ytho
print(text[:4])    # Pyth
print(text[2:])    # thon
print(text[:])     # full string
```

```python
text = "Python"

print(text[0:6:2])   # Pto
print(text[::-1])    # reverse string
```

# String Immutability

Strings are immutable, which means they cannot be changed after creation.

```python
text = "Python"

# This will cause an error
# text[0] = "J"
```

# Basic String Operations

```python
a = "Hello"
b = "World"

# Concatenation
print(a + " " + b)

# Repetition
print(a * 3)
```
