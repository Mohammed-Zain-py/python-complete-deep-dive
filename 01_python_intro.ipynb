# What is Programming?

Programming is the process of **writing instructions for a computer to perform specific tasks**.

These instructions are written using a **programming language**.

A program can be used to:

* Solve problems
* Automate tasks
* Process data
* Build applications
* Control systems

In simple terms:

Programming = **Giving step-by-step instructions to a computer to solve a problem**

# What is Python?

Python is a **high-level, object-oriented, interpreted, scripting language as well as programming language** known for its **simplicity and readability**.

It was created by **Guido van Rossum** and released in **1991**.

Python is designed to be easy to learn and easy to use, making it one of the most popular programming languages in the world.

# Why Python is Popular

Python is widely used because of the following reasons:

* Easy to learn and read
* Large community support
* Huge number of libraries
* Works on multiple platforms
* Suitable for beginners and professionals

# Applications of Python

Python is used in many fields including:

* Web Development
* Data Science
* Machine Learning
* Artificial Intelligence
* Automation
* Game Development
* Cyber Security

# Python Modules

A **module** is a file containing Python code such as functions, variables, and classes.

Modules allow us to **reuse code instead of writing everything from scratch**.

Python provides many built-in modules such as:

* math
* random
* datetime
* os

# Python Comments

Comments are used to **explain code and make programs easier to understand**.

Python ignores comments while executing the program.

Two types of comments are commonly used:

* Single-line comments
* Multi-line comments

```python
# This is a single line comment

print("Hello World")

# Python ignores comments during execution
```

```python
"""
This is a multi-line comment or
docstring that can span across
multiple lines.
"""

print("Learning Python step by step")
```
