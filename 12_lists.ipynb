# What are Lists?

A list is a **collection of items stored in a single variable**.

Lists are:

* Ordered
* Mutable (can be changed)
* Allow duplicate values

```python
numbers = [1, 2, 3, 4, 5]
names = ["Zain", "John", "Alice"]

print(numbers)
print(names)
```

# Indexing in Lists

Each element has an index starting from 0.

```python
numbers = [10, 20, 30, 40]

print(numbers[0])  # 10
print(numbers[2])  # 30
```

# Negative Indexing

Negative indexing starts from -1.

```python
numbers = [10, 20, 30, 40]

print(numbers[-1])  # 40
print(numbers[-2])  # 30
```

# Accessing Elements Using Loop

```python
numbers = [1, 2, 3, 4]

for num in numbers:
    print(num)
```

# Basic List Operations

```python
a = [1, 2, 3]
b = [4, 5]

# Concatenation
print(a + b)

# Repetition
print(a * 2)
```

# Checking Element in List

```python
numbers = [1, 2, 3, 4]

print(2 in numbers)
print(5 in numbers)
```

# Practice Problems

```python
# Problem 1: Print all elements of a list

numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print(num)
```

```python
# Problem 2: Find sum of elements

numbers = [1, 2, 3, 4, 5]

total = 0
for num in numbers:
    total += num

print(total)
```

```python
# Problem 3: Find maximum element

numbers = [10, 5, 20, 8]

print(max(numbers))
```

```python
# Problem 4: Count number of elements

numbers = [1, 2, 3, 4]

print(len(numbers))
```

```python
# Problem 5: Check if element exists

numbers = [1, 2, 3, 4]

print(3 in numbers)
```
