# What are Loop Control Statements?

Loop control statements are used to **control the flow of loops**.

They help in:

* Stopping a loop early
* Skipping certain iterations
* Executing code after loop completion

# break Statement

The `break` statement is used to **exit the loop immediately**.

Once break is executed, the loop stops completely.

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

# continue Statement

The `continue` statement is used to **skip the current iteration** and move to the next one.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

# else with Loop

The `else` block runs **only if the loop completes normally** (without break).

```python
for i in range(1, 6):
    print(i)
else:
    print("Loop finished successfully")
```

```python
for i in range(1, 6):
    if i == 3:
        break
    print(i)
else:
    print("This will not execute")
```

# Practice Problems

```python
# Problem 1: Stop printing when number reaches 7

for i in range(1, 11):
    if i == 7:
        break
    print(i)
```

```python
# Problem 2: Skip multiples of 3

for i in range(1, 11):
    if i % 3 == 0:
        continue
    print(i)
```

```python
# Problem 3: Check if a number is prime

num = 7

for i in range(2, num):
    if num % i == 0:
        print("Not Prime")
        break
else:
    print("Prime")
```

```python
# Problem 4: Find first number divisible by 5

for i in range(1, 20):
    if i % 5 == 0:
        print(i)
        break
```
