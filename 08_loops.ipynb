# What are Loops?

Loops are used to **repeat a block of code multiple times**.

They help avoid writing the same code again and again.

Example:

Print numbers from 1 to 5 → using loop instead of writing 5 print statements.

# for Loop

The `for` loop is used to iterate over a sequence (like list, string, range).

Syntax:

for variable in sequence:
    code

```python
for i in range(5):
    print(i)
```

```python
for i in range(1, 6):
    print(i)
```

# range() Function

The `range()` function is used to generate a sequence of numbers.

Forms:

range(stop)
range(start, stop)
range(start, stop, step)

```python
print(list(range(5)))          # 0 to 4
print(list(range(1, 6)))       # 1 to 5
print(list(range(1, 10, 2)))   # step of 2
```

# Looping Through a String

```python
text = "Python"

for char in text:
    print(char)
```

# while Loop

The `while` loop runs as long as a condition is True.

Syntax:

while condition:
    code

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

# Practice Problems

```python
# Problem 1: Print numbers from 1 to 10

for i in range(1, 11):
    print(i)
```

```python
# Problem 2: Print even numbers from 1 to 10

for i in range(1, 11):
    if i % 2 == 0:
        print(i)
```

```python
# Problem 3: Print sum of first 5 numbers

total = 0

for i in range(1, 6):
    total += i

print(total)
```

```python
# Problem 4: Print multiplication table of 5

num = 5

for i in range(1, 11):
    print(num * i)
```

```python
# Problem 5: Reverse a string using loop

text = "Python"
rev = ""

for char in text:
    rev = char + rev

print(rev)
```
